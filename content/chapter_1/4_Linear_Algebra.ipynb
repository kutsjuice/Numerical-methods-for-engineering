# Основы линейной алгебры на Julia
## Скалярное произведение векторов
Cтоит заметить, что хотя обозначение скалярного произведение при помощи $\left<\cdot\right>$ является общепринятым и хорошо знакомо читателю из курса линейной алгебры, довольно часто можно  повстречать запись следующего вида
\begin{equation}
    x^T y
    \tag{1}
\end{equation}
Данная запись подразумевает обращение с векторами как с матрицами содержащими один столбец. В этом случае операция транспонирования сделает из него матрицу, содержащую одну строку. Запись (1) будет абсолютно эквивалента привычной
$$ x \cdot y $$
